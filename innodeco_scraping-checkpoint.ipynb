```python
import requests
import os
from pathlib import Path

for index, row in df.iterrows():
    # Define the directory path using both type and color
    directory_path = Path(f"data/{row['Category']}/{row['Product_Type']}/{row['Product_Color']}")
    
    # Create the directory if it does not exist
    os.makedirs(directory_path, exist_ok=True)
    
    # Define the file path
    file_path = directory_path / f"{row['Product_ID']}.jpg"
    
    # Download and save the image
    try:
        image = requests.get(row['Product_URL'])
        with open(file_path, 'wb') as f:
            f.write(image.content)
        print(f"Saved {file_path}")
    except Exception as e:
        print(f"Error downloading {row['Product_URL']}: {e}")
```
